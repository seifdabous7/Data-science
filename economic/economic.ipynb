```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('Economic Data - 9 Countries (1980-2020).csv')
```

```python
df = data.copy()
```

```python
df.info()
```

```python
df.groupby('country')['gdppercent'].mean().sort_values(ascending=False)
```

```python
df.groupby('country')['gdppercent'].max().sort_values(ascending= False)
```

```python
befour_2008 = df[(df['year'])>=2005 & (df['year']<=2008)]
befour_mean = befour_2008.groupby('country')['gdppercent'].mean().sort_values(ascending=False)
```

```python
after_2008 = df[(df['year']>=2008)&(df['year']<=2010)]
after_mean= after_2008.groupby('country')['gdppercent'].mean().sort_values(ascending=False)
```

```python
gdb_comparison = pd.DataFrame({
    'gdb_befour_2008': befour_mean,
    'gdb_after_2008': after_mean
})
```

```python
gdb_comparison
```

```python
inflation_stability =(df.groupby('country')['gdppercent'].std().sort_values)
```

```python
inflation_stability
```

```python
corr_by_country = (
    df.groupby("country")
      .apply(lambda x: x["gdppercent"].corr(x["inflationrate"]))
      .sort_values(ascending=False)
)

corr_by_country
```

```python
gdp_unemp_corr = (
    df.groupby("country")
      .apply(lambda x: x["gdppercent"].corr(x["unemploymentrate"]))
      .sort_values(ascending=False)
)

gdp_unemp_corr
```
