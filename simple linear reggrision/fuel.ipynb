```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('fuel.csv')
```

```python
df = data.copy()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.scatterplot(data= df,x='Horse Power',y='Fuel Economy (MPG)')
plt.title('the relation between hourse power and mpg')
plt.xlabel('hourse power')
plt.ylabel('mpg')
plt.show()
```

```python
sns.regplot(data =df,x = 'Horse Power',y= 'Fuel Economy (MPG)')
plt.show()
```

```python
x = df[['Horse Power']]
y = df['Fuel Economy (MPG)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```
