```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('random_fitness_dataset.csv')
```

```python
df = data.copy()
```

```python
plt.Figure(figsize=(8,6))
sns.scatterplot(data=df,x='age', y='steps', hue='gender')
plt.title('relation between age and steps')
plt.xlabel('age')
plt.ylabel('steps')
```
