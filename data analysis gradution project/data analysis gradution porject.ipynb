```python

def fill_missing_increment_id(df):
    
    df['increment_id'] = df['increment_id'].astype(str)
    
    missing_mask = df['increment_id'].isnull() | (df['increment_id'] == 'nan')
    missing_count = missing_mask.sum()
    
    if missing_count > 0:
        print(f"  missing values increment_id: {missing_count}")
        
        existing_ids = df.loc[~missing_mask, 'increment_id']
        if not existing_ids.empty:
            numeric_ids = pd.to_numeric(existing_ids, errors='coerce')
            max_id = numeric_ids.max()
            
            start_id = int(max_id) + 1
            new_ids = [str(start_id + i) for i in range(missing_count)]
            
            df.loc[missing_mask, 'increment_id'] = new_ids
            
            print(f" New IDs created from {start_id} to {start_id + missing_count - 1}")
    
    return df

df = fill_missing_increment_id(df)
```
