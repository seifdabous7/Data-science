```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('StudentsPerformance.csv')
```

```python
df = data.copy()
```

```python
plt.Figure(figsize=(8,6))
sns.scatterplot( data=df,x='math score',y='reading score',hue='gender')
plt.title('relation between reading and math')
plt.xlabel('math score')
plt.ylabel('reading')
plt.show()
```

```python
fig1 = px.scatter(df,x='math score',y='reading score', color='gender')
fig1
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df,x='race/ethnicity',y='math score')
plt.title('race for math')
plt.xlabel('race')
plt.ylabel('math score')
plt.show()
```

```python
fig2 = px.bar(df,x='race/ethnicity',y='math score')
```

```python
fig2
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(df['math score'])
plt.title('math score')
plt.xlabel('math score')
plt.ylabel('fequance')
plt.show()
```

```python
fig3 = px.histogram(df,x = 'math score')
```

```python
fig3
```

```python
plt.Figure(figsize=(8,6))
sns.boxplot(data=df,x='gender',y='math score')
plt.title('realtion between gender and math score')
plt.xlabel('gender')
plt.ylabel('math score')
plt.show()
```
