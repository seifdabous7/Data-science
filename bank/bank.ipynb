```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('bank.csv')
```

```python
df = data.copy()
```

```python
df['age'].mean()
```

```python
df.info()
```

```python
bins = [0, 19, 29, 39, 49, 59, 69, 100]
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']

df['AgeGroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)

age_group_counts = df.groupby('AgeGroup')['customer_id'].count()

print("\nNumber of customers in each age group:")
print(age_group_counts)

most_common_group = age_group_counts.idxmax()
print(f"\nThe most common age group is: {most_common_group}")
```
