```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy import stats
from scipy.special import inv_boxcox

```

```python
np.random.seed(0)
sales_data = np.random.exponential(scale=100, size=1000)
```

```python
df = pd.DataFrame(sales_data, columns=['sales'])
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.Figure(figsize=(8,6))
plt.hist(df['sales'],bins=30,density=True,color='blue')
plt.title('destribution of sales')
plt.xlabel('sales')
plt.ylabel('destripution')
plt.show()
```

```python
df['sales'],lam = stats.boxcox(df['sales'])
```

```python
plt.Figure(figsize=(8,6))
plt.hist(df['sales'],bins=30,density=True,color='blue')
plt.title('destribution of sales')
plt.xlabel('sales')
plt.ylabel('destripution')
plt.show()
```

```python
df['origanal_sales'] = inv_boxcox(df['sales'],lam)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df['sales'] = scaler.fit_transform(df[['sales']])
```

```python
plt.Figure(figsize=(8,6))
plt.hist(df['sales'],bins=30,density=True,color='blue')
plt.title('destribution of sales')
plt.xlabel('sales')
plt.ylabel('destripution')
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
power_transfromer = PowerTransformer(method='yeo-johnson')
```

```python
df['sales']= power_transfromer.fit_transform(df[['sales']])
```

```python
plt.Figure(figsize=(8,6))
plt.hist(df['sales'],bins=30,density=True,color='blue')
plt.title('destribution of sales')
plt.xlabel('sales')
plt.ylabel('destripution')
plt.show()
```
